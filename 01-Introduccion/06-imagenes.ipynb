```python
import numpy as np
import random
import matplotlib.pyplot as plt
from scipy import ndimage
from scipy import misc

rows = 10
columns = 10
height = 1080
width = 1920

#mapache = misc.face()
mapache = plt.imread("/home/tkhacker/Pictures/Pictures/vocaloid_hatsune_miku_schoolgirl_player_bags_bridge_window_102550_1920x1080.jpg")

#DIVISIÓN POR COLUMNAS
mapache_splited = np.hsplit(mapache, columns)
random.shuffle(mapache_splited)
mapache_columns = np.array(mapache_splited)
mapache_columns = np.swapaxes(mapache_columns, 0, 1)
mapache_columns = mapache_columns.reshape(height, width, 3)

#DIVISIÓN POR FILAS
mapache_splited = np.vsplit(mapache_columns, rows)
random.shuffle(mapache_splited)
mapache_rows = np.array(mapache_splited)
mapache_rows = mapache_rows.reshape(height, width, 3)

plt.imshow(mapache_rows)
plt.show()
```
